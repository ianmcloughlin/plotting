```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("ggplot")
```

```python
plt.plot([1,2,3], "r-", label="Red line")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3], "r-", label="Red line")
ax.legend()
```

#### From the seaborn tutorial

***

```python
x = np.linspace(0.0, 100.0, 40)
y = x**2 + np.random.normal(0, 1000, len(x))
```

```python
sns.relplot(x=x, y=y);
```

<br>

#### Distplots

***

```python
x = np.random.normal(1, 2, 1000)
```

```python
sns.displot(x, kde=True)
```

#### Bar chart

```python
makes = ["Toyota", "Skoda", "VW", "BMW"]
cars = np.random.choice(makes, 10000)
list(cars)
```

```python
sns.countplot(x=cars)
```
